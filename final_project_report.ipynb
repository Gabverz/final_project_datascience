# Relatório final

Conforme plano de trabalho inicial, as seguintes etapas foram desenvolvidas:

1 - Criação de documento orientativo: documento apresentando objetivo, motivação, contexto geral da atividade, análise exploratória de dados e plano de resolução geral do projeto;

2 - Tratamento dos dados: tratamento dos dados que serão utilizados como features e target da análise preditiva, como também criação de dataframe base para aplicação dos métodos;

3 - Treinamento do modelo: aplicação de modelos preditivos de regressão como LightGBM ou CatBoost para o conjunto de dados;

4 - Validação dos modelos: aplicação do modelo treinado em conjunto de validação;

5 - Conclusão: conclusões extraídas e potencial de ganho calculado.

### 1 - Criação de documento orientativo

A etapa 01 foi executada e tinha o intuito de apresentar ao problema e principais dados envolvidos, como também analisá-los a partir da análise exploratória. Foram constatados poucos problemas com os dados nesta etapa, porém ficou claro a necessidade de trabalhar os valores ausentes e as métricas temporais de alguma forma antes de seguir com o desenvolvimento do modelo.

Conhecimento técnico acerca do problema facilitou a compreensão dos dados, visto que algumas variáveis fisicamente não influenciam a solução proposta, como as datas das leituras. Isto favoreceu o modelo, pois menos dados tiveram que ser tratados e lidos para determinar a temperatura final do processo.

Ficou pendente na etapa 1 realizar uma análise de outliers e do comportamento gráfico das variáveis, porém foi escolhido realizar na etapa seguinte pois foi onde algumas variáveis foram criadas e a tabela-base de dados consolidada.

### 2 - Tratamento dos dados

Foram realizados os tratamentos de dados de cada dataframe nesta etapa. As variáveis consideradas essenciais ao problema foram criadas a partir dos dados originais, utilizando principalmente o método .groupby(). Para os valores de energia de cada período ('energy_total') encontrados, foi utilizazda a equação abaixo:

$$Q_{efetivo} = P_{ativa} \times ({t_{end} - t_{start}}) $$

Ao gerar todos os dados estatísticos, condensamos tudo em um só dataframe, onde era possível extrair para cada lote dezenas de informações úteis ao modelo.

Em seguida, foi aplicada uma análise visual dos dados para identificar o tipo de comportamento dos mesmos e se havia muitos outliers. A análise identificou uma falha nos dados, onde havia uma potência negativa dentre os valores. Isto não faz sentido dentro da amostragem e do sistema em questão, portanto foi necessário excluir dados negativos do DF antes de seguir com a construção do modelo. Ao utilizar o método .merge() com 'outer', alguns valores de temperatura final (variável target) estavam em branco, o que também foi corrigido.

Ainda foi aplicado um filtro de scaling para normalização dos dados, buscando melhor fit entre eles. Entretanto, isto provou-se inútil, tendo em vista que lidamos com um problema de regressão. Os modelos foram treinados com e sem normalização e geraram os mesmos resultados.

Com o dataframe base criado, seguimos para o treinamento dos modelos.

### 3 / 4 - Treinamento e validação dos modelos

Foram escolhidos os modelos CatBoost e LightGBM para este desenvolvimento, pois ambos conseguem lidar com valores nulos, têm custo computacional baixo e conseguem convergir muito bem dentro dos dados que dispomos.

Foram separados 20% dos dados para validação. Foi avaliados utilizar 25 ou 30%, entretanto o banco de dados não é robusto o suficiente então é mais seguro acertar num modelo de treinamento mais arrojado e validação com menos dados.

O modelo CatBoost foi treinado com otimização de parâmetros apenas para o número de iterações, enquanto o LighGBM foi otimizado para 'leaves' e 'rounds'. Ambos, ao final do treinamento, conseguiram resultados muito significativos, com R² = 0,95 e REQM menor que 4,0 para os dois casos, enquanto que o modelo padrão utilizando a média dos valores de temperatura atingiu REQM de 18,2. 

### 5 - Conclusão

No presente projeto, foram desenvolvidos dois modelos de aprendizado de máquina com base em algoritmos de gradiente descendente para previsão de dados de temperatura final do processo de produção de aço em lingotes.

- **Resultados obtidos**

Os modelos treinados foram capazes de atingir erro absoluto médio de apenas 2.7 °C. Dada a média dos valores de temperatura igual a 1593.9 °C, isso representa um desvio de 1,6% em média. O R² de ambos os modelos chegaram a valores também muito interessantes, acima de 0.95, o que é considerado excelente para previsões em escala industrial de processos físico-químicos. Desta forma, acredita-se que o modelo pode prever com precisão valores de temperatura a partir de dados previamente fornecidos. Vale ressaltar também que as bibliotecas utilizadas para modelagem são de alta eficiência e consideradas modernas, portanto não requerem um custo computacional muito alto.

Visto que as unidades não foram fornecidas, ganhos energéticos não podem ser mensurados. Entretanto, um caminho para otimização do processo produtivo pode ser via aplicação de séries de valores para cada variável, entendendo assim seu impacto no valor de temperatura final. Ficou claro pela análise de importância que as características principais são as temperaturas anteriores, o que faz sentido pois a última temperatura nunca irá se distanciar demais das outras. 

Desta forma, pode-se concluir que o projeto entregou o resultado desejado, baseado em modelos precisos de previsão dos mais modernos atualmente, que puderam entregar resultados significativos para o sistema em estudo. O modelo final a ser adotado deve ser o LightGBM, que apresentou R² = 0,95 e REQM = 4,0, similar ao CatBoost, porém com custo computacional ligeiramente melhor.
